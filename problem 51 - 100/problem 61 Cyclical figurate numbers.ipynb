```python
# problem 61 Triangle, square, pentagonal, hexagonal, heptagonal, and octagonal numbers are all figurate (polygonal) numbers 
#and are generated by the following formulae:

#Triangle	 	P3,n=n(n+1)/2	 	1, 3, 6, 10, 15, ...
#Square	 		P4,n=n2 			1, 4, 9, 16, 25, ...
#Pentagonal	 	P5,n=n(3n−1)/2	 	1, 5, 12, 22, 35, ...
#Hexagonal	 	P6,n=n(2n−1)	 	1, 6, 15, 28, 45, ...
#Heptagonal	 	P7,n=n(5n−3)/2	 	1, 7, 18, 34, 55, ...
#Octagonal	 	P8,n=n(3n−2)	 	1, 8, 21, 40, 65, ...
#The ordered set of three 4-digit numbers: 8128, 2882, 8281, has three interesting properties.

#The set is cyclic, in that the last two digits of each number is the first two digits of the next number 
#(including the last number with the first).
#Each polygonal type: triangle (P3,127=8128), square (P4,91=8281), and pentagonal (P5,44=2882), is represented by a 
#different number in the set.
#This is the only set of 4-digit numbers with this property.
#Find the sum of the only ordered set of six cyclic 4-digit numbers for which each polygonal type: 
#triangle, square, pentagonal, hexagonal, heptagonal, and octagonal, is represented by a different number in the set.
```

```python
trilist = []
sqlist = []
pentlist = []
hexlist = []
heptlist = []
octlist = []
n = 1
reqdlen = 4
while(True):
    tribool = sqbool = pentbool = hexbool = heptbool = octbool = False
    
    trival = str(int(n*(n+1)/2))
    lenval = len(str(trival)) 
    if(lenval <= reqdlen):
        if(lenval == reqdlen):
            trilist.append(trival)
        tribool = True

    sqval = str(int(n*n))
    lenval = len(str(sqval)) 
    if(lenval <= reqdlen):
        if(lenval == reqdlen):
            sqlist.append(sqval)
        sqbool = True
    
    pentval = str(int(n*(3*n - 1)/2))
    lenval = len(str(pentval))
    if(lenval <= reqdlen):
        if(lenval == reqdlen):
            pentlist.append(pentval)
        pentbool = True
        
    hexval = str(int(n*(2*n - 1)))
    lenval = len(str(hexval))
    if(lenval <= reqdlen):
        if(lenval == reqdlen):
            hexlist.append(hexval)
        hexbool = True
        
    heptval = str(int(n*(5*n - 3)/2))
    lenval = len(str(heptval))
    if(lenval <= reqdlen):
        if(lenval == reqdlen):
            heptlist.append(heptval)
        heptbool = True
        
    octval = str(int(n*(3*n - 2)))
    lenval = len(str(octval))
    if(lenval <= reqdlen):
        if(lenval == reqdlen):
            octlist.append(octval)
        octbool = True
        
    n += 1
    if(tribool or sqbool or pentbool or hexbool or heptbool or octbool):
        continue
    else:
        break

import itertools as it
permutes = list(it.permutations(list(range(6))))

polylists = {
    0: trilist,
    1: sqlist,
    2: pentlist,
    3: hexlist,
    4: heptlist,
    5: octlist
}
```
