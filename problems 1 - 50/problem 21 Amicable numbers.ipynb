```python
#problem 21 Let d(n) be defined as the sum of proper divisors of n (numbers less than n which divide evenly into n).
#If d(a) = b and d(b) = a, where a â‰  b, then a and b are an amicable pair and each of a and b are called amicable numbers.

#For example, the proper divisors of 220 are 1, 2, 4, 5, 10, 11, 20, 22, 44, 55 and 110; therefore d(220) = 284. The proper divisors of 284 are 1, 2, 4, 71 and 142; so d(284) = 220.

#Evaluate the sum of all the amicable numbers under 10000.
```

```python
import math
def generatedivisors(num, div):
    i = 1
    while i <= math.sqrt(num): 
          
        if (num % i == 0) : 
              
            # If divisors are equal, print only one 
            if (num / i == i) : 
                div.append(i) 
            else : 
                # Otherwise print both 
                if(i == 1):
                    div.append(i)
                else:
                    div.append(i) 
                    div.append(int(num/i)) 
        i = i + 1
    return div

#generate divisors of all the numbers
data = {}
divisors = []
for i in range(1, 10001):
    data[i] = {}
    data[i]['divisors'] = generatedivisors(i, divisors)
    data[i]['add'] = sum(data[i]['divisors'])
    divisors = []
    
aminos = []    
for i in range(1, 10001):
    if(data[i]['add'] < 10000):
        if(data[data[i]['add']]['add'] == i and data[i]['add'] != i):
            aminos.append(i)
            aminos.append(data[i]['add'])
            
aminos = list(set(aminos))
print(sum(aminos))
```
