```python
#problem 15 Starting in the top left corner of a 2×2 grid, and only being able to move to the right and down, there are exactly 6 routes to the bottom right corner.


#How many such routes are there through a 20×20 grid?
```

```python
import numpy as np

gridsize = 21
grid = np.zeros([gridsize , gridsize], dtype = 'int64')

for i in range(0, gridsize -1):
    grid[i][gridsize-1] = 1
    grid[gridsize-1][i] = 1

count = gridsize-2

while(count >=0):
    for i in range(count, -1, -1):
        grid[i][count] = grid[i+1][count] + grid[i][count+1]
        grid[count][i] = grid[i+1][count] + grid[i][count+1]
    count -= 1
grid
```
