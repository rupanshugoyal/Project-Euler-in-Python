```python
# problem 12 The sequence of triangle numbers is generated by adding the natural numbers. So the 7th triangle number would be 1 + 2 + 3 + 4 + 5 + 6 + 7 = 28. The first ten terms would be:

#  1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ...

#Let us list the factors of the first seven triangle numbers:

 1: 1
 3: 1,3
 6: 1,2,3,6
10: 1,2,5,10
15: 1,3,5,15
21: 1,3,7,21
28: 1,2,4,7,14,28
#We can see that 28 is the first triangle number to have over five divisors.

#What is the value of the first triangle number to have over five hundred divisors?
```

```python
import math
def generatedivisors(num, div):
    i = 1
    while i <= math.sqrt(num): 
          
        if (num % i == 0) : 
              
            # If divisors are equal, print only one 
            if (num / i == i) : 
                div.append(i) 
            else : 
                # Otherwise print both 
                div.append(i) 
                div.append(num/i) 
        i = i + 1
    return div

count = 1
divisors = []
trino = 0
while(len(divisors) <= 500):
    divisors = []
    trino = int(count*(count + 1)/2)
    count += 1
    divisors = generatedivisors(trino, divisors)
    
print(trino)
```
