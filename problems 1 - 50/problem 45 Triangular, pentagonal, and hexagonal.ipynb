```python
#problem 45 Triangle, pentagonal, and hexagonal numbers are generated by the following formulae:

#    Triangle	 	Tn=n(n+1)/2	 	1, 3, 6, 10, 15, ...
#    Pentagonal	 	Pn=n(3n−1)/2	 	1, 5, 12, 22, 35, ...
#    Hexagonal	 	Hn=n(2n−1)	 	1, 6, 15, 28, 45, ...
#It can be verified that T285 = P165 = H143 = 40755.

#Find the next triangle number that is also pentagonal and hexagonal.
```

```python
trilist = []
penlist = []
hexlist = []
i = 2
while(True):
    trilist.append(int(i*(i+1)/2))
    penlist.append(int(i*(3*i-1)/2))
    hexlist.append(int(i*(2*i-1)))
    if(trilist[i-2] in penlist and trilist[i-2] in hexlist):
        if(trilist[i-2] != 40755):
            print(trilist[i-2])
            break
    i+=1
```
